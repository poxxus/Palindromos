```python
dia, mes, ano = input(), input(), input()
dia, mes, ano = [int(x) for x in [dia, mes, ano]]
mesesde31, mesesde30, fevereiro = False, False, False
if mes in (1,2,4,6,8,9,11):
   mesesde31 = True
if mes in (5,7,10,12):
   mesesde30 = True
if mes==3:
   fevereiro = True   
#Meses precedidos por um mês de 31 dias      
if mesesde31 == True:
     dia = dia - 1
     if dia<1:
        dia = 31
        mes = mes -1
        if mes<1:
           mes = 12
           ano = ano - 1 
     print(dia,mes,ano)
##Meses precedidos por um mês de 30 dias
if mesesde30 == True:
      dia = dia - 1
      if dia<1:
         dia = 30
         mes = mes - 1
         if mes<1:
            mes = 12
            ano = ano - 1          
      print(dia,mes,ano)
#Mês de março com o caso do bissexto                    
if fevereiro==True:
    if ano%4==0 and ano%100!=0 or ano%400==0:
        dia = dia - 1
        if dia<1:
          dia = 29
          mes = mes - 1
        print(dia,mes,ano)
    else:
        dia = dia - 1
        if dia<1:
         dia = 28
         mes = mes - 1
        print(dia,mes,ano)
```
