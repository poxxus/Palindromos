```python
A = float(input("Altura em metros: ")) 
P = float(input("Peso em kg: "))
AA = A*A
IMC = P/AA
if IMC<16:
    print("Magreza grave")
elif 16<=IMC<17:
    print("Magreza moderada")
elif 17<=IMC<18.5:
    print("Magreza leve")
elif 18.5<=IMC<25:
    print("Saudavel")
elif 25<=IMC<30:
    print("Sobrepeso")
elif 30<=IMC<35:
    print("Obesidade Grau I")
elif 35<=IMC<40:
    print("Obesidade Grau II")
else:
    print("Obesidade Grau III")

```
