```python
C, P, B = int(input("Quilowatts-hora consumidos: ")), float(input("Preço de cada quilowatt-hora: ")), int(input("Bandeira tarifária vigente: "))
if B == 1: ##Bandeira verde
  A = 0
elif B == 2: ##Bandeira Amarela
  A = 0.015
elif B == 3:  ##Bandeira Vermelha 1
  A = 0.030
else: ##Bandeira Vermelha 2
  A = 0.0449
R = C*(P+A)
print("{:.2f}".format(R))
```
